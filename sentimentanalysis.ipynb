```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
import flair
```

```python
sentiment_model =flair.models.TextClassifier.load('en-sentiment')
```

```python
df = pd.read_csv('C:/Users/Jessie Angelica/OneDrive/Documents/Alibaba task/infovesta/sentiment.csv')
df.head()
```

```python
df.drop_duplicates(subset='released', keep='first', inplace=True)
df.head()
```

```python
sample = 'hello world!'
sample = flair.data.Sentence(sample)
```

```python
sentiment_model.predict(sample)
sample
```

```python
sample.labels
```

```python
sample.labels[0].value
sample.labels[0].score
```

```python
sentiment = []
confidence = []

for sentence in df['url']:
    if sentence .strip()== "":
        sentiment.append("")
        confidence.append("")
    else:
        sample =flair.data.Sentence(sentence)
        sentiment_model.predict(sample)

        sentiment.append(sample.labels[0].value)
        confidence.append(sample.labels[0].score)
```

```python
sample
```

```python
df['sentiment'] = sentiment
df['Confidence'] = confidence
```
